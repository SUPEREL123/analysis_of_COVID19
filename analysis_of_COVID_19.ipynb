```python
lowest_100_countries = total_counts_entity.nsmallest(50)
plt.figure(figsize=(8, 8))
plt.pie(lowest_100_countries, labels=lowest_100_countries.index, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.title('Total Number of Lowest 50 Countries')
plt.show()
```

## 2. pie chart of confirmed deaths due to covid-19

```python
df_Africa = df[df['Entity']=="Africa"]
df_Asia = df[df['Entity']=="Asia"]
df_Europe = df[df['Entity']=="Europe"]
df_North_America = df[df['Entity']=="North America"]
df_South_America = df[df['Entity']=="South America"]
df_Oceania = df[df['Entity']=="Oceania"]
df_Antarctica = df[df['Entity']=="Antarctica"]



```

```python
# Add a 'Continent' column to each DataFrame to identify them after concatenation
df_Africa['Continent'] = 'Africa'
df_Asia['Continent'] = 'Asia'
df_Europe['Continent'] = 'Europe'
df_North_America['Continent'] = 'North America'
df_South_America['Continent'] = 'South America'
df_Oceania['Continent'] = 'Oceania'
df_Antarctica['Continent'] = 'Antarctica'

# Combine all DataFrames into one
df_all = pd.concat([df_Africa, df_Asia, df_Europe, df_North_America, df_South_America, df_Oceania, df_Antarctica])

# Plot the data
plt.figure(figsize=(10, 10))

# Loop through each continent and plot its data
for continent, data in df_all.groupby('Continent'):
    plt.plot(data['Day'], data['Total confirmed deaths due to COVID-19 per 100,000 people'], label=continent)

# Add labels, title, and legend
plt.xlabel('Day')
plt.ylabel('Total Confirmed Deaths per 100,000 People')
plt.title('Total Confirmed COVID-19 Deaths per 100,000 People by Continent')
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

## Africa has the lowest number of death. I think that's not because they control the virus better than other continent but they did not report the death to the government problem or issue

## South America has the most death in COVID-19

## Europe and North America has similar pattern shown in the graph

## when other continent keep increasing the number of death, Oceania keep very stable, I think that's because Oceania is the only continent that is in the South half of the Earth.

```python
total_counts = df_all.groupby('Continent')['Total confirmed deaths due to COVID-19 per 100,000 people'].sum()


plt.figure(figsize=(10, 10))
plt.pie(total_counts, labels=total_counts.index, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.title('Total confirmed deaths due to COVID-19 per 100,000 people in seven continents')
plt.show()
```
